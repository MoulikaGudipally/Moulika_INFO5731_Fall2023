# Question 1

(20 points). Write a Python program to generate 12 random numbers between 1 and 100. Sort the list of numbers in ascending order and print it out. Sort the list of numbers in descending order and print it out. Calculate the average for the group and print it out.

```python
# Your code here
import random

# Generate 12 random numbers between 1 and 100
random_numbers = [random.randint(1, 100) for _ in range(12)]

# Sort the list in ascending order
ascending_sorted = sorted(random_numbers)

# Sort the list in descending order
descending_sorted = sorted(random_numbers, reverse=True)

# Calculate the average of the numbers
average = sum(random_numbers) / len(random_numbers)

# Print the results
print("Generated Numbers:", random_numbers)
print("Ascending Sorted:", ascending_sorted)
print("Descending Sorted:", descending_sorted)
print("Average:", average)


```

# Question 2
(25 points). Write a program that will do the following string formatting tasks:

(1) Ask the user to enter two sentences.

(2) Make each sentence into a list. Each element in the list should correspond to a word in the sentence. For example, if the sentence is "Alas! Am I alive? This is my first python program", the program should print out ['Alas, '!', 'Am', 'I', 'alive, '?', 'This', 'is', 'my', 'first', 'python', 'program'].

(3) Compare the two sentences. Print out a list of words that occur in both sentences.

(4) Remove any punctuation from the sentence and print it back out. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”)

(5) Count the number of vowels in the sentences. Print out each vowel and the number of times it appears in the sentences, such as: a:2, e:1, i:0, etc

```python
# Your code here
import string

# (1) Ask the user to enter two sentences.
sentence1 = input("Enter the first sentence: ")
sentence2 = input("Enter the second sentence: ")

# (2) Make each sentence into a list of words.
def sentence_to_word_list(sentence):
    # Split the sentence into words
    words = sentence.split()

    # Separate punctuation from words
    word_list = []
    for word in words:
        stripped_word = word.strip(string.punctuation)
        if stripped_word:
            punctuation = ''.join(char for char in word if char in string.punctuation)
            word_list.extend([stripped_word, punctuation])

    return word_list

word_list1 = sentence_to_word_list(sentence1)
word_list2 = sentence_to_word_list(sentence2)

# (3) Compare the two sentences and print out words that occur in both.
common_words = set(word_list1).intersection(word_list2)

# (4) Remove punctuation from the sentences and print them.
def remove_punctuation(sentence):
    translator = str.maketrans('', '', string.punctuation)
    return sentence.translate(translator)

sentence1_no_punctuation = remove_punctuation(sentence1)
sentence2_no_punctuation = remove_punctuation(sentence2)

# (5) Count the number of vowels in the sentences.
vowels = 'aeiouAEIOU'

def count_vowels(sentence):
    vowel_count = {vowel: 0 for vowel in vowels}
    for char in sentence:
        if char in vowels:
            vowel_count[char.lower()] += 1
    return vowel_count

vowel_count1 = count_vowels(sentence1)
vowel_count2 = count_vowels(sentence2)

# Print the results
print("(2) Word lists:")
print("Sentence 1:", word_list1)
print("Sentence 2:", word_list2)
print("(3) Common words:", common_words)
print("(4) Sentences without punctuation:")
print("Sentence 1:", sentence1_no_punctuation)
print("Sentence 2:", sentence2_no_punctuation)
print("(5) Vowel counts:")
print("Sentence 1:", ', '.join(f'{v}: {count}' for v, count in vowel_count1.items()))
print("Sentence 2:", ', '.join(f'{v}: {count}' for v, count in vowel_count2.items()))

```

Write a program to implement the following questions.

(1) (10 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(2) (10 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(3) (10 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(4) (10 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
import csv

# Function to calculate the difference in 'value' from year to year
def calculate_difference(data):
    differences = {}

    for i in range(1, len(data)):
        year1 = int(data[i - 1]['year'])
        year2 = int(data[i]['year'])
        value1 = int(data[i - 1]['value'])
        value2 = int(data[i]['value'])
        difference = value2 - value1
        differences[f'{year1}-{year2}'] = difference

    # Print the differences in ascending order
    for year_diff, diff in sorted(differences.items(), key=lambda x: x[1]):
        print(f'{year_diff}: {diff}')

# Function to calculate and print the percentages of vacant and occupied housing units
def calculate_percentages(data):
    print('Year   Vacant Housing Units   Occupied Housing Units')
    for entry in data:
        year = entry['year']
        total_units = int(entry['value'])
        vacant_units = int(entry['value'])
        occupied_units = total_units - vacant_units
        vacant_percentage = (vacant_units / total_units) * 100
        occupied_percentage = (occupied_units / total_units) * 100
        print(f'{year}   {vacant_percentage:.2f}%   {occupied_percentage:.2f}%')

# Function to find the year with the highest number of vacant and occupied housing units
def find_highest_units(data):
    max_vacant_units = max(data, key=lambda x: int(x['value']))
    min_vacant_units = min(data, key=lambda x: int(x['value']))

    print('Year    Value')
    print(f'Vacant Housing Units    {max_vacant_units["year"]}    {max_vacant_units["value"]}')
    print(f'Occupied Housing Units  {min_vacant_units["year"]}    {min_vacant_units["value"]}')

# Function to calculate and print the harmonic mean of total housing units
def calculate_harmonic_mean(data):
    total_units = [int(entry['value']) for entry in data]
    n = len(total_units)
    harmonic_mean = n / sum(1 / x for x in total_units)

    print(f'Harmonic Mean of Total Housing Units: {harmonic_mean:.2f}')

# Main function
if __name__ == '__main__':
    with open('Assignment1_denton_housing.csv', newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        data = list(reader)

    print('Task 1:')
    calculate_difference(data)

    print('\nTask 2:')
    calculate_percentages(data)

    print('\nTask 3:')
    find_highest_units(data)

    print('\nTask 4:')
    calculate_harmonic_mean(data)
```
